## Two options for subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10, 5))

# PLot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplot option 2
fig, ax = plt.subplots(ncols=2,
                      nrows=2,
                      figsize=(10, 5))
# Plot to each different index
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

### Which one should I use? (pyplot vs matplotlib OO method?)
* When plotting something quickly, okay to use pyplot
* When plotting something more advanced, use OO method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
# Pyplot method
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target');
```

```python
# OO method combinated with pyplot method
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);

#ax.set_xlim([45, 100])
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"]);

# Costumize the plot
ax.set(title = "Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel='Cholesterol');

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle='--');
```

```python
over_50.head()
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows = 2,
                              ncols = 1,
                              figsize = (10, 10),
                              sharex=True);

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

# Costumize ax0
ax0.set(title = "Heart Disease and Cholesterol Levels",
      ylabel='Cholesterol');

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target");

# Add a mean line
ax0.axhline(over_50["chol"].mean(),
          linestyle='--');

# Add data to ax1
scatter2 = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"])

# Costumize ax0
ax1.set(title = "Heart Disease and Max Heart Rate",
      xlabel="Age",
      ylabel="Max Heart Rate");


# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")

# Add a mean line
ax1.axhline(over_50["thalach"].mean(),
           linestyle='--')

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python

# OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                    cmap="winter"); # this changes the collor scheme

# Costumize the plot
ax.set(title = "Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel='Cholesterol');

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target", labelcolor="black", edgecolor="black");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle='--');
```

```python
#Costumizing the X and y axis limitations
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows = 2,
                              ncols = 1,
                              figsize = (10, 10),
                              sharex=True);

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap = "winter")

# Costumize ax0
ax0.set_xlim([50, 80])
ax0.set(title = "Heart Disease and Cholesterol Levels",
      ylabel='Cholesterol');

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target");

# Add a mean line
ax0.axhline(over_50["chol"].mean(),
          linestyle='--');

# Add data to ax1
scatter2 = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"],
                      cmap = "summer")

# Costumize ax1
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])
ax1.set(title = "Heart Disease and Max Heart Rate",
      xlabel="Age",
      ylabel="Max Heart Rate");


# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")

# Add a mean line
ax1.axhline(over_50["thalach"].mean(),
           linestyle='--')

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
